```python
#  Q8
# P(mxm) B(mxn) QT(nxn)
P, B, QT = bidiagonalise(A)

print("\n============================================================")
print("[Q8]")
print("\nP =\n", P)
print("\nB =\n", B)
print("\nQT =\n", QT)
print("\n============================================================")


# 1. reconstruction of A:
# check A ?= P B QT
print("\n[Q8][1]Check whether A ≈ P @ B @ QT\n")

A_hat = P @ B @ QT
print("A_hat = P @ B @ QT = \n", A_hat)
print("\nA_hat ?= A :", np.allclose(A_hat, A))
print("\n============================================================\n")

# 2. check the orthogonality of P and Q
print("[Q8][2]check the orthogonality of P and Q")

PT_P = P.T @ P
print("\nPT_P =\n", PT_P)
print("\nPT_P ?= I :", np.allclose(PT_P, np.eye(m)))

QT_Q = QT @ QT.T
print("\nQT_Q =\n", QT_Q)
print("\nQT_Q ?= I :", np.allclose(QT_Q, np.eye(n)))
print("\n============================================================\n")

# 3. check whether B is upper bidiagonal
print("[Q8][3]check whether B is upper bidiagonal")
print("\nDefinition: bij = 0 unless j ∈ {i, i + 1}\n")

def is_upper_bidiagonal(B, tol=1e-6):
    m, n = B.shape
    for i in range(m):
        for j in range(n):
            if j != i and j != i+1 and abs(B[i, j]) > tol:
                return False
    return True


print("Is B upper bidiagonal? :", is_upper_bidiagonal(B))
print("\n============================================================\n")
```
